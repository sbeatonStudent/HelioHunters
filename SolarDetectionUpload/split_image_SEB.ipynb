```python
from PIL import Image
import random

random_number = random.randint(0, 100)


def split_image(image_path, mask_path, output_train_directory, output_test_directory):
    # Open the image
    imageNGB = Image.open(image_path)
    imageMask = Image.open(mask_path)

    # Get the width and height of the image
    width, height = imageNGB.size

    # Calculate the number of grid squares
    num_squares = (width // 128) * (height // 128)

    # Iterate over the grid squares
    count = 1
    for i in range(0, width, 128):
        for j in range(0, height, 128):
            # Extract the grid square as a new image
            squareNGB = imageNGB.crop((i, j, i + 128, j + 128))
            squareMask = imageMask.crop((i, j, i + 128, j + 128))

            random_number = random.randint(0, 100)

            if random_number>80:
                output_directory_image = output_test_directory
                output_directory_mask = output_test_mask_directory
            else:
                output_directory_image = output_train_directory
                output_directory_mask = output_train_mask_directory

            # Save the grid square as a separate PNG file
            output_path = f"{output_directory_image}/i_{count}.png"
            squareNGB.save(output_path)
            # Save the grid square as a separate PNG file
            output_path = f"{output_directory_mask}/m_{count}.png"
            squareMask.save(output_path)
            print(f"Saved grid square {count} as {output_path}")

            count += 1
            print(count)

# Example usage
image_path = "data/IRGBoutput.png"
mask_path = "data/mask.png"
output_train_directory = "dataset/train/images"
output_test_directory = "dataset/test/images"
output_train_mask_directory = "dataset/train/masks"
output_test_mask_directory = "dataset/test/masks"

split_image(image_path, mask_path, output_train_directory, output_test_directory)
```
